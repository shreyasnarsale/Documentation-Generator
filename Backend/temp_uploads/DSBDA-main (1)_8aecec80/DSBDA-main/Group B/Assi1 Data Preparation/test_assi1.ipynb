```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("dataset_Facebook.csv",sep=";")
data
```

```python
data.isnull().sum()
```

```python
number = data.select_dtypes(include='number')
mv = number.mean()
data = data.fillna(mv)
data
```

```python
data.isnull().sum()
```

```python
data.columns
```

# Create a Datasubsets

```python
s1 = data[data["like"] > 100]
s1
```

```python
s2 = data[data["Type"] == "Photo"]
s2
```

```python
data = data[data['Post Weekday'] < 6]
data
```

```python
s4 = data[["Lifetime Post Total Impressions","like","share","comment"]]
s4
```

```python
s5 = data[["Lifetime Post Total Impressions","like","Category","Type"]]
s5
```

```python
s6 = data.iloc[:100]
s6
```

```python
s7 = data.loc[:,['Type','like']]
s7
```

# Merge Data

```python
# Row wise
merge_row_wise = pd.concat([s1,s2], axis = 0)
merge_row_wise
```

```python
#Merge_Column_wise
merge_column_wiese = pd.merge(s4,s5, on = ["Lifetime Post Total Impressions","like"], how = 'left')
merge_column_wiese
```

# Sort

```python
sort = data.sort_values(by = 'like',ascending = True)
sort
```

```python
sort = data.sort_values(by = 'like',ascending = False)
sort
```

# Shape and Reshape

```python
data.shape
```

```python
melt_data = data.melt(id_vars = ['Type']  , value_vars = ['like','share','comment'])
melt_data
```

```python
melted_data = melt_data.drop_duplicates(subset=['Type', 'variable'])
melted_data
```

```python
pivoted_data = melted_data.pivot(index='Type', columns='variable', values='value')
pivoted_data
```
