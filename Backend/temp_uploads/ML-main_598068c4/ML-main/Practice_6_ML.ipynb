```python
import csv

data = []
with open('Market_Basket.csv') as file:
  reader = csv.reader(file,delimiter=',')
  for row in reader:
      data +=[row]
```

```python
from apyori import apriori
results = list(apriori(data,min_support=0.01,min_confidence=0.4,min_lift=1.0))
results
```

```python
len(results)
```

```python
rules = []
supports = []
confidences = []
```

```python
# Extract rules , support, confidences from the results

for result in results:
  support = result.support
  for rule in result.ordered_statistics:
    antacedent = tuple(rule.items_base)
    consequent = tuple(rule.items_add)
    rules.append(f'{antacedent} => {consequent}')
    supports.append(support)
    confidences.append(rule.confidence)

# Create a dataframe to extracted data

df_results = pd.DataFrame(data={'Rules':rules,'Support':supports,'Confidence':confidences})
df_results
```
